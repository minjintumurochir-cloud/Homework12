```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate dataset with 20 records
df = pd.DataFrame({
    'name': ['Peter', 'Peter', 'Tom', 'Tom', 'Tom', 'John',
             'Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank',
             'Grace', 'Henry', 'Ivy', 'Jack', 'Kate', 'Leo',
             'Mike', 'Nina'],
    'website': [
        'https://aa.com.tw/web1/web2/web3',
        'http://aa.com.tw/web5/web3/web6',
        'http://aa.com.tw/web1/web2/web4/web2',
        'http://aa.com.tw/web7/web8/web5/web1',
        'http://aa.com.tw/web6/web5/web9/web2/web3',
        'http://aa.com.tw/web1/web2/web5/web2',
        'http://aa.com.tw/web1/page1',
        'http://aa.com.tw/web2/page2',
        'http://aa.com.tw/web3/page3',
        'http://aa.com.tw/web4/page4',
        'http://aa.com.tw/web5/page5',
        'http://aa.com.tw/web6/page6',
        'http://aa.com.tw/web7/page7',
        'http://aa.com.tw/web8/page8',
        'http://aa.com.tw/web9/page9',
        'http://aa.com.tw/web10/page10',
        'http://aa.com.tw/web1/web2/web3',
        'http://aa.com.tw/web2/web3/web4',
        'http://aa.com.tw/web1/web2/web3',
        'http://aa.com.tw/web5/web6/web7'
    ]
})

# Display dataset
print("Dataset (20 records):")
print(df)
print(f"\nTotal records: {len(df)}")
```

```python
# Question 1: How many browsers (unique users)?
num_browsers = df['name'].nunique()
print(f"Question 1: How many browsers? {num_browsers}")
print(f"\nList of users: {list(df['name'].unique())}")
```

```python
# Question 2: How many unique web pages?
num_pages = df['website'].nunique()
print(f"Question 2: How many web pages? {num_pages}")
```

```python
# Question 3: Who are the most loyalty clients?
loyalty_clients = df['name'].value_counts()
print("Question 3: Most loyalty clients (number of visits):")
print(loyalty_clients)
print(f"\nMost loyal client: {loyalty_clients.index[0]} with {loyalty_clients.values[0]} visits")
```

```python
# Question 4: What are the most frequent browsed web pages?
frequent_pages = df['website'].value_counts()
print("Question 4: Most frequently browsed web pages:")
print(frequent_pages.head(10))
print(f"\nMost visited page: {frequent_pages.index[0]}")
print(f"Number of visits: {frequent_pages.values[0]}")
```

```python
# Create visualizations
fig, axes = plt.subplots(2, 1, figsize=(12, 10))

# Chart 1: Loyalty clients
loyalty_clients.plot(kind='bar', ax=axes[0], color='skyblue', edgecolor='black')
axes[0].set_title('Loyalty Clients (Number of Visits per User)', fontsize=14, fontweight='bold')
axes[0].set_xlabel('User Name', fontsize=12)
axes[0].set_ylabel('Number of Visits', fontsize=12)
axes[0].grid(axis='y', alpha=0.3)

# Chart 2: Top 10 most visited pages
frequent_pages.head(10).plot(kind='barh', ax=axes[1], color='lightcoral', edgecolor='black')
axes[1].set_title('Top 10 Most Frequently Visited Pages', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Number of Visits', fontsize=12)
axes[1].set_ylabel('Website URL', fontsize=12)
axes[1].grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Final summary report
print("\n" + "="*60)
print("           FINAL SUMMARY REPORT")
print("="*60)
print(f"Total Records: {len(df)}")
print(f"Unique Users (Browsers): {num_browsers}")
print(f"Unique Web Pages: {num_pages}")
print(f"Most Loyal Client: {loyalty_clients.index[0]} ({loyalty_clients.values[0]} visits)")
print(f"Most Visited Page: {frequent_pages.index[0]}")
print(f"Most Visited Page Count: {frequent_pages.values[0]} visits")
print("="*60)
```
